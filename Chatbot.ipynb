```python
import json
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, GlobalAveragePooling1D
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
```

```python
Sequential?
```

```python
with open('intents.json') as file:
    data = json.load(file)
```

```python
train_sentance = []
train_tag = []
lables = []
response = []
for intent in data['intents']:
    for pattern in intent['patterns']:
        train_sentance.append(pattern)
        train_tag.append(intent['tag'])
    response.append(intent['responses'])
    
    if intent['tag'] not in lables:
        lables.append(intent['tag'])
        
num_class = len(lables)
```

```python
print(num_class)
print(lables)
print(train_sentance)
```

```python
le = LabelEncoder()
train_tag = le.fit_transform(train_tag)
```

```python
tokenizer = Tokenizer(num_words=1000, oov_token="<OOV>")
tokenizer.fit_on_texts(train_sentance)
word_index = tokenizer.word_index
sequence_token = tokenizer.texts_to_sequences(train_sentance)
pad_sequences_token = pad_sequences(sequence_token, truncating='post', maxlen=20)
```

```python
#print(pad_sequences_token)
```

```python
model = Sequential()
model.add(Embedding(1000,16, input_length= 20))
model.add(GlobalAveragePooling1D())
model.add(Dense(16, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(num_class, activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy',optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(pad_sequences_token, np.array(train_tag), epochs= 500)
```

```python
model.save('chat_model')
```
